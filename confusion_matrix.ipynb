```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
#creates confusion matrix

knownAVPs=np.ones(293)
known_nonAVPS= np.zeros(293)
#input true positive (TP), false negative (FN), true negative (TN), and false positive (FP) values. 
TP=np.ones(249)
FN= np.zeros(44)
TN=np.zeros(237)
FP=np.ones(56)

y_true = np.concatenate((knownAVPs,known_nonAVPS))  # true labels
y_pred = np.concatenate((TP,FN,TN,FP))  # predicted labels

# Computes confusion matrix
conf_matrix = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Predicted Negative', 'Predicted Positive'], yticklabels=['Actual Negative', 'Actual Positive'])
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix for AVP Model')
plt.show()
```

```python
ones=np.ones(293)
zeros = np.zeros(293)
TP=246
FN=47
TN=245
FP=48
TPdat=np.ones(TP)
FNdat= np.zeros(FN)
TNdat=np.zeros(TN)
FPdat=np.ones(FP)

y_true = np.concatenate((ones,zeros))  # true labels 
y_pred = np.concatenate((TPdat,FNdat,TNdat,FPdat))  # predicted labels 

# Compute the confusion matrix
conf_matrix = confusion_matrix(y_true, y_pred)

#calculate confusion matrix stats
accuracy = (TP+TN)/float(TP+FP+TN+FN)
precision = TP / float(TP + FP)
sensitivity = TP / float(FN + TP)
specificity = TN / float(TN + FP)
MCC=((TP*TN)-(FP*FN))/float(np.sqrt((TP + FP)*(TP + FN)*(TN + FP)*(TN + FN)))

# Plot the confusion matrix using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Greens', xticklabels=['Predicted Negative', 'Predicted Positive'], yticklabels=['Actual Negative', 'Actual Positive'])
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix for AVP+GAN Model')
plt.show()
{'accuracy':accuracy,
'precision':precision,
'sensitivity':sensitivity,
'specificity':specificity,
'mcc':MCC}
```
